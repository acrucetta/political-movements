```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Compile count data
a = pd.read_csv("../data/NYT/NYT_Authoritarianism.csv")
b = pd.read_csv("../data/NYT/NYT_Communism.csv")
c = pd.read_csv("../data/NYT/NYT_Conservatism.csv")
d = pd.read_csv("../data/NYT/NYT_Fascism.csv")
e = pd.read_csv("../data/NYT/NYT_Liberalism.csv")
f = pd.read_csv("../data/NYT/NYT_Libertarianism.csv")
g = pd.read_csv("../data/NYT/NYT_Marxism.csv")
h = pd.read_csv("../data/NYT/NYT_Nationalism.csv")
i = pd.read_csv("../data/NYT/NYT_Populism.csv")
j = pd.read_csv("../data/NYT/NYT_Progressivism.csv")
k = pd.read_csv("../data/NYT/NYT_QAnon.csv")
l = pd.read_csv("../data/NYT/NYT_Socialism.csv")

# Compile NER data
a1 = pd.read_csv("../data/NYT/NYT_Authoritarianism_NER.csv")
b1 = pd.read_csv("../data/NYT/NYT_Communism_NER.csv")
c1 = pd.read_csv("../data/NYT/NYT_Conservatism_NER.csv")
d1 = pd.read_csv("../data/NYT/NYT_Fascism_NER.csv")
e1 = pd.read_csv("../data/NYT/NYT_Liberalism_NER.csv")
f1 = pd.read_csv("../data/NYT/NYT_Libertarianism_NER.csv")
g1 = pd.read_csv("../data/NYT/NYT_Marxism_NER.csv")
h1 = pd.read_csv("../data/NYT/NYT_Nationalism_NER.csv")
i1 = pd.read_csv("../data/NYT/NYT_Populism_NER.csv")
j1 = pd.read_csv("../data/NYT/NYT_Progressivism_NER.csv")
k1 = pd.read_csv("../data/NYT/NYT_QAnon_NER.csv")
l1 = pd.read_csv("../data/NYT/NYT_Socialism_NER.csv")
```

```python
data_comb = pd.concat([a,b,c,d,e,f,g,h,i,j,k,l], ignore_index=True)
ner_data_comb = pd.concat([a1,b1,c1,d1,e1,f1,g1,h1,i1,j1,k1,l1], ignore_index=True)
```

```python
data_comb_yr = data_comb[['Movement','Count','Year']].groupby(['Movement','Year']).sum().reset_index()
```

```python
plt.figure(figsize=(16, 10))
sns.lineplot(data_comb_yr['Year'], data_comb_yr['Count'], hue=data_comb_yr['Movement']).set(
    title='Mentions in NYT Articles by Movement by Year')
```

```python
plt.figure(figsize=(16, 10))
data_comb_yr2 = data_comb_yr[(data_comb_yr.Movement != 'Conservatism') & (data_comb_yr.Movement != 'Liberalism')]
sns.lineplot(data_comb_yr2['Year'], data_comb_yr2['Count'], hue=data_comb_yr2['Movement']).set(
    title='Mentions in NYT Articles by Movement by Year')
```

```python
data_comb.to_csv("NYT_Count_Comb.csv")
```
